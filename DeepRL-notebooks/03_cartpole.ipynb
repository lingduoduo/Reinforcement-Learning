```python
import gym
```

```python
env = gym.make("CartPole-v0", render_mode="human")
```

```python
state = env.reset()
```

```python
for t in range(1000):
    env.render()
    print(state)
    
    # a random action
    action = env.action_space.sample()
    
    # implement action to change enviorments, observed state, reward, decided termited or truncated and info.
    state, reward, terminated, truncated, info = env.step(action)

    if terminated or truncated:
        print("Finished")
        state = env.reset()
```

```python
env.close()
```
