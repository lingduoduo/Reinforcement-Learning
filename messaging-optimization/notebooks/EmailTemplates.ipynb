```python
from bs4 import BeautifulSoup
import json
```

```python
jsonData = json.load(open("20180423-crm-discover-weekly-reactivation-copy"))
htmlText = jsonData["data"]["html"]["int"]["html"]
soup = BeautifulSoup(htmlText, 'html.parser')
```

```python
print(soup.title.text)
```

```python
for tag in soup.find_all(True):
    if(tag.name == "img"):
        print(tag.get(key="src"))
```
